Least Squares Linear Regression

```python
# Example from the textbook.

import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['figure.figsize'] = [6, 6]
plt.rcParams.update({'font.size': 12})

#Data Matrix A, and RHS vector b
#Heat of cement vs. mix of ingredients
#Original data from Matlab, https://www.mathworks.com/help/stats/sample-data-sets.html
#https://github.com/dynamicslab/databook_python/tree/master/DATA

A = np.genfromtxt('https://raw.githubusercontent.com/kursatkara/MAE_5020_S24/master/Data/hald_ingredients.csv', delimiter=',')
b = np.genfromtxt('https://raw.githubusercontent.com/kursatkara/MAE_5020_S24/master/Data/hald_heat.csv', delimiter=',')
```

```python
print("Ingredients")
print(A.shape, A.size)
print(A)
```

```python
print("Heat generated for each experiment")
print(b.shape, b.size)
print(b)
```

```python
 # Solve Ax=b using SVD
U, S, VT = np.linalg.svd(A,full_matrices=False)
x = VT.T @ np.linalg.inv(np.diag(S)) @ U.T @ b
```

```python
# True (Given Data)
plt.plot(b, color='k', linewidth=1.5, label='Heat Data')

# Regression using SVD
plt.plot(A@x, '-o', color='r', linewidth=1.5, markersize=6, label='Regression')
plt.grid(linestyle='--')
plt.legend()
plt.show()
```
