Find the eigenvalues and eigenvectors of a positive definite orhogonal matrix and plot the eigenvectors.

```python


import numpy as np
import matplotlib.pyplot as plt

define_own_matrix = True

if define_own_matrix:
  # if you chose to define your own 2 by 2 matrix
  A = np.array([[2.7348836, -1.98235045], [-1.98235045, 3.2651164 ]])

else:
  # Steps 1-3 is to create a positive definite orthogonal matrix.
  # You may define your own matrix and skip these steps.

  # Step 1: Create a random 2x2 matrix and apply QR decomposition and
  # obtain an orthonormal (orthogonal) matrice.
  #
  # Background: Orthogonal matrices are matrices in which the columns vectors form
  # an orthonormal set (each column vector has length one and is orthogonal
  # to all the other colum vectors). For square orthonormal matrices,
  # the inverse is simply the transpose, Q-1 = QT.
  Q, _ = np.linalg.qr(np.random.rand(2, 2))

  # Step 2: Create a diagonal matrix with positive eigenvalues
  eigenvalues = np.array([1.0, 5.0])

  # Step 3: Compute the positive definite orthogonal matrix
  A = Q @ np.diag(eigenvalues) @ Q.T
  print("Orthogonal Matrix, Q :", Q)



# Step 4: Find the eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(A)

# Step 5: Plot the eigenvectors
origin = [0, 0]

fig, ax = plt.subplots()
ax.quiver(*origin, *eigenvectors[:,0], angles='xy', scale_units='xy', scale=1, color='r', label='Eigenvector 1')
ax.quiver(*origin, *eigenvectors[:,1], angles='xy', scale_units='xy', scale=1, color='b', label='Eigenvector 2')

ax.set_xlim(-1, 1)
ax.set_ylim(-1, 1)
ax.set_aspect('equal')
ax.legend()

plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.title('Eigenvectors of the Positive Definite Orthogonal Matrix')
plt.show()

#print("Matrix A :", A)
#print("Eigenvalues of A:", eigenvalues)
#print("Eigenvectors of A:", eigenvectors)
```
